# Parte 1

## 1.1 Variables

Los Dataframe que analizaremos contienen variables de tipo númerico continuo. 

El período de tiempo del analisis va desde el 1 de enero de 2018 hasta el 16 de mayo de 2021, pero no todas las crypto tienen los datos completos para ese período.


```python
for i in cryptos.keys():

  print('La crypto '+str(i)+' tiene datos entre el '+str(cryptos[i][cryptos[i]['open'].notnull()].index.min().date())+
        ' y el '+str(cryptos[i][cryptos[i]['open'].notnull()].index.max().date()))
```

## 1.2 Valores faltantes.

```python
valores_criptos = dict()
for i in cryptos.keys():
  first_date = cryptos[i].first_valid_index()
  empty_rows = cryptos[i].query('index >= @first_date').open.isna().sum()
  total_rows = len(cryptos[i].query('index >= @first_date').open)
  valores_criptos[i] = total_rows
  print('Para la crypto '+str(i)+' hay '+str(empty_rows)+' filas sin datos sobre un total de '+str(total_rows))
```

Para tratar los valores faltantes tenemos distintas opciones:


*   Eliminar las filas con datos faltantes
*   Imputar usando métodos de series de tiempo como forward fill, back fill o interpolación lineal

Si tuvieramos que elegir una, preferiríamos imputar usando interpolación lineal, ya que no consideramos conveniente eliminar las filas y al no saber si el valor faltante sería mayor o menor que sus valores contiguos la interpolación es un método más preciso. 

## 1.3 Distribuciones

```python
cryptos_daily = dict()
for i in cryptos.keys():
  cryptos_daily[i] = cryptos[i].resample('D').agg({'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last', 'volume': 'sum'})
```

```python
for i,c in enumerate(cryptos_daily.keys()):
  p = sns.displot(data=cryptos_daily[c], x='close', kde=True)
  plt.title(c)
  plt.xlabel('Valor al cierre')
```

```python
# Generamos un dataframe con cada cripto y su media, desvío y coeficiente de variación mensual para el último año
coin = list(cryptos_daily.keys())
medidas = ['mean', 'std', 'cv']
# Date range junio 2020 - junio 2021
idx = pd.date_range(start='2020-06-01', end='2021-6-01', freq='M', closed='left')
cols = list(itertools.product(coin,medidas))
ml_index= pd.MultiIndex.from_tuples(cols, names=['cripto', 'metrica'])
df_cripto_metrics = pd.DataFrame(columns=ml_index, index=idx)
df_cripto_metrics.head(2)
```

```python
# Llenamos el dataframe con las medidas estadísticas.
def cv(x): return np.std(x)/np.mean(x)
for i in coin:
    temp = cryptos[i].query('index > @idx.min()').resample('M').agg({'close':['mean', 'std', cv]}).values
    df_cripto_metrics[i] = temp
df_cripto_metrics
```

## 1.4 Outliers

Para calcular los outliers usaremos las siguientes definiciones:
- outlier < Q1 - 1.5*IQR
- outlier > Q3 + 1.5*IQR
Siengo Q1 y Q3 el cuantil 25 y 75 respectivamente y IQR el rango inter cuartil

```python
def cuantiles(x):
    return x.quantile(0.25)[0], x.quantile(0.75)[0]
```

```python
for i in cryptos.keys():
    c = cryptos[i][['close']]
    q1,q3 = cuantiles(c)
    lim1 = q1 - 1.5*(q3-q1)
    lim2 = q3 + 1.5*(q3-q1)
    outliers = len(c.query('close < @lim1')) + len(c.query('close > @lim2'))
    print(f'Para la cripto {i} hay en total {outliers} outliers de un total de {valores_criptos[i]}')
```

Posteriormente, cuando analizamos el comportamiento de los criptoactivos a lo largo del tiempo, vimos que muchos de ellos tuvieron subidas abruptas de precios, lo cual explicaría la alta cantidad de outliers en algunas de ellas.

## 1.7 Evolución temporal


Primero unificamos todos los precios de cierre en un mismo dataframe para facilitar su visualización mediante seaborn.

Visto que los cryptoactivos tienen valores muy diferentes expresados en dólares (Por ejemplo un bitcoin a principios de mayo de 2020 valía alrededor de 50000 dólares mientras que la unidad de cardano cotizaba en alrededor de 1,5) decidimos escalar la variable para facilitar su comparación. 

Esto implica que el precio de cierre de todas las monedas pasará a estar entre 0 y 1, ya que no nos interesa su valor absoluto sino más bien su comportamiento.

```python
#Unifico variables en un dataframe
closing_prices = pd.DataFrame(index=cryptos_daily['BTCUSDT'].index)

for i in cryptos_daily.keys():
  closing_prices[i] = cryptos_daily[i].close

#Libreria para escalar
from sklearn.preprocessing import MinMaxScaler

#Escalo variables
for i in list(closing_prices):
  closing_prices[i] = MinMaxScaler().fit_transform(closing_prices[[i]])
```

En el primer gráfico analizamos los 4 cryptoactivos con mayor capitalización de mercado.

```python
plt.subplots(figsize=(28,8))
sns.lineplot(data=closing_prices[['BTCUSDT','ETHUSDT','BNBUSDT','ADAUSDT']], dashes=False)
```

```python
plt.subplots(figsize=(28,8))
sns.lineplot(data=closing_prices[['BTCUSDT','ETCUSDT','BNBUSDT','ADAUSDT']], dashes=False)
```

En el análisis de las principales monedas vemos que durante 2018 su valor fue disminuyendo, después un 2019 de relativa estabilidad (con mayores aumentos en el bitcoin) que se interrumpió con la llegada de la pandemia de COVID en el 2020 (ver la bajada abrupta en marzo-abril) y después hacia fines de 2020 empezó una carrera alcista que se mantuvo hasta el final del período, aunque con eventuales correcciones de precios.
